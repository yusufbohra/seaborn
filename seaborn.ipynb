```python
#que 3
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_dataset = sns.load_dataset("titanic")

# Plot box plots using "pclass" as x-axis and "age" and "fare" as y-axes
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_dataset, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_dataset, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#que 4
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_dataset = sns.load_dataset("diamonds")

# Plot histogram with 'price' column and 'cut' column as hue
sns.histplot(data=diamonds_dataset, x='price', hue='cut', kde=True)

# Add title and labels
plt.title("Histogram of Price by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Show the plot
plt.show()
```
